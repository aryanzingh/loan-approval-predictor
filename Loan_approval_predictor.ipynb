```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/loan_approval_dataset.csv")
df.columns=df.columns.str.strip()
df['education']
```

```python
df['self_employed'].unique()
```

```python
df['education']=df['education'].replace({' Graduate':1,' Not Graduate':0})
```

```python
df['self_employed']=df['self_employed'].replace({' No':0,' Yes':1})
```

```python
df
```

```python
df=df.drop('loan_id',axis=1)
```

```python
df
```

```python
Y_label=df['loan_status'].map({' Approved':1,' Rejected':0})
```

```python
df['total_assets']=df['bank_asset_value']+df['commercial_assets_value']+df['residential_assets_value']+df['luxury_assets_value']

df=df.drop(['commercial_assets_value','residential_assets_value','luxury_assets_value','bank_asset_value'],axis=1)
df
```

```python
df['loan_income_ratio']=df['loan_amount']/df['income_annum']
```

```python
df['emi']=df['loan_amount']/df['loan_term']
```

```python
X_label=df.drop('loan_status',axis=1).values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_label,Y_label,test_size=0.2,random_state=1)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression()
mod.fit(X_train,Y_train)
```

```python
mod.score(X_test,Y_test)
```

```python
import joblib
from google.colab import files

joblib.dump(mod,    "model.pkl");   files.download("model.pkl")
joblib.dump(scaler, "scaler.pkl");  files.download("scaler.pkl")
```
